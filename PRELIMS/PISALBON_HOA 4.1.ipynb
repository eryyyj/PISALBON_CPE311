# Hands-on Activity 4.1: The Tower of Hanoi Problem

```python
class TowerofHanoi:
    def __init__(self):
        pass
    def Move(self,Pegs,sour,aux,dest):
        if Pegs == 1:
            print("move disk", Pegs,' from', sour , 'to', dest)
        else:
            self.Move(Pegs-1,sour,dest,aux)
            print("move disk", Pegs,' from', sour , 'to', dest)
            self.Move(Pegs-1,aux,sour,dest)
```

```python
a = True
while a:
    response = int(input("====================\n"
                         '1 make a move\n'
                         '0 exit\n'
                         "====================\n"
                         'input: '))
    
    match response:
        case 1:
            g = TowerofHanoi()
            numberOfPegs = int(input("Enter the number of pegs: "))
            g.Move(numberOfPegs,'A','B','C')
        case 0:
            a = False
            print('NOW EXITING')
        case _:
            print('INVALID CHOICE!')
    
```
