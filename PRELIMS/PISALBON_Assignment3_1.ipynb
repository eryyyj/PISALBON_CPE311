##The Solution:


the base case of an unconnected graph is the vertices is 12. Vertices greater or equal to 12 are can be considered as unconnected graph by following the graph representation below image.png

the parameters of the graph solution must be the compnent must have at least 6 vertices in it the code representation of the graph is as follows:


```
self.graph = [{1:{2},2:{1,3,4},3:{2,4,5},4:{2,3,6},5:{3},6:{4}},{7:{8},8:{7,9,10},9:{8,10,11},10:{8,9,12},11:{9},12:{10}}]
```



vertices greater 12 are connected in sequence from node 12


visual representation: image.png

Code representation:




````python
class Graph:
  def __init__(self):
    #generate the graph solution
    self.graph = [{1:{2},2:{1,3,4},3:{2,4,5},4:{2,3,6},5:{3},6:{4}},
     {7:{8},8:{7,9,10},9:{8,10,11},10:{8,9,12},11:{9},12:{10}}]
  def unconnectedgraphChecker(self,vertices):
    #conditional statement for the base case of connected graph:
    if vertices < 12 and vertices >= 1:
      return "No"
    else: #else statement if the vertices is 12 and above
      # conditional statement for the base case of connected graph:
      if vertices == 12:
        print('component 1:',self.graph[0])
        print('component 2:', self.graph[1])
        return "Yes"
      elif vertices > 12 and vertices <= 100: #else statement if the vertex is greater than 12
        for i in range(12,vertices):
          self.add_edge({i,i+1})
        print('component 1:',self.graph[0])
        print('component 2:', self.graph[1])
        return "Yes"
      else:
        return "Input must be between 1 and 100!"
  def add_edge(self, edge):
    edge = set(edge)
    vertex1,vertex2 = tuple(edge)
    for x,y in [(vertex1,vertex2), (vertex2,vertex1)]:
      #if statement if the node is already in the graph
      if x in self.graph[1]:
        self.graph[1][x].add(y)
      else:# else statement if the node is not in the graph
        self.graph[1][x] = {y}

graph = Graph()
inputt = int(input("Enter a number of nodes: "))
graph.unconnectedgraphChecker(inputt)
````
