```python
import uuid 
import random

class Splitwise:
    def __init__(self,grp_name):
        self.members_dict = {} #members list
        self.grp_name = grp_name
        print('group has been added!') #prompt for successfully created a group
        self.v = 1 # iterator for the id
    def get_groupName(self): #getters for the name of the group
        return self.grp_name
    def __str__(self):
        return f"Group Name: {self.grp_name} , Members: {self.members_dict}"
    def addMembers(self): #method for adding a member for a specific group
        print("============================")
        n = int(input('Enter the number of members you want to add\n'
                      'input: ')) 
        for i in range(n):
            print("============================")
            memName = input('Enter the name of the member\n'
                           'input: ')
            if memName not in self.members_dict:
                self.members_dict[self.v] = [memName,0]
                print('successfully added the member!')
            else:
                print('member already added please add a unique name!')
                continue
            self.v += 1
        return "all members have been added!!"
    def transactions(self):
        print("============================")
        amount_left = 0 # tracks how much that the payers have payed and will split to the members
        n = len(self.members_dict)
        m = int(input('Enter the number of transactions: '))
        balance = [0 for i in range(n)]# creates a list of 0 in nth elements depending on the number of members
        print(balance)
        for t in range(m): # code for making a transaction
            print('transaction_id: ', random.randint(1, 1000))
            np,ns = map(int,input('Enter the number of payers and the number of splits: ').split()) 
            """ input for the number of payers 
                and the number of splits np = # of payers, ns = # of splits  """

            for i in range(np): # code for making a payment
                memid,amt = map(int,input('Enter the payer id and the amount of money you will give them: ').split()) 
                """input for the payer id and the amount paid 
                memid = payerID, amount = amount paid by the payer"""
                self.members_dict[memid][1] -= amt # updating the debt and balance on the members list
                balance[memid - 1] -= amt
                amount_left -= amt
                
            a = ns # no of splits tracker
            for i in range(ns): # no of splits tracker
                """input for the payer id and the amount paid 
                memid = splitterID, amt = amount to be split"""
                print('you have ', a ,' split/s left.')
                print('amount of bill that you can give splits: ', abs(amount_left))
                memid,amt = map(int,input('Enter the member id and the amount they will split: ').split())
                self.members_dict[memid][1] += amt # updating the debt and balance on the members list
                balance[memid - 1] += amt
                a -= 1
                amount_left += amt

        ans = []
        j = 0
        for i in range(n): # loop for generating the 
            if(balance[i] > 0): # executes if the member has a balance debt
                cur_bal = balance[i] # copy the balance debt amount
                while (cur_bal > 0 and j<n):
                    if(balance[j]>=0): # conditional statement in determining who is the payer
                        j = j+1;
                        continue;
                    minimum_bal = min(cur_bal, abs(balance[j]))
                    cur_bal -= minimum_bal #checking if the amount payed is same as the amount split
                    balance[j] += minimum_bal # updating the current amount that the payer has left
                    ans.append([[i+1, j+1], minimum_bal])

        for i in range(len(ans)):
            print(ans[i][0][0],ans[i][0][1],ans[i][1])
```
